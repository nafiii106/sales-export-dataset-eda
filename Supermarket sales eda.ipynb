## Revenue analysis

```python
# the total revenue, total cost and total profit for the entire dataset

df.columns = df.columns.str.strip()
df['order_value_EUR'] = pd.to_numeric(df['order_value_EUR'].astype(str).str.replace(',',''), errors='coerce')
df["profit"] = df["order_value_EUR"] - df["cost"]
print(df[['order_value_EUR','cost','profit']].sum().round(2))
```

```python
# country generates the highest total revenue & the highest average order value

df.groupby('country')['order_value_EUR'].agg(['sum','mean']).sort_values('sum', ascending = False).head(1)
```

```python
# product category has the highest profit margin on average

df.groupby('category')['profit'].mean().sort_values(ascending = False).head(1)
```

```python
# the top 5 sales managers by total revenue and by total profit

df.groupby('sales_manager')[["profit",'order_value_EUR']].sum().sort_values("order_value_EUR",ascending = False).head(5)
```

```python
# sales rep closes the highest average order value

df.groupby('sales_rep')['order_value_EUR'].agg(['mean','count']).sort_values('mean', ascending=False).head(10)
```
